# Apply our assumption about CSUSHPISA groups by using K-means Clustering

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'DATE' is in datetime format
data['DATE'] = pd.to_datetime(data['DATE'])
data['DATE_numeric'] = data['DATE'].astype(np.int64) // 10**9  # Convert to Unix timestamp in seconds

# Fit KMeans for different numbers of clusters
Distortion = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(data[['CSUSHPISA', 'DATE_numeric']])
    Distortion.append(kmeans.inertia_)

# Plot Elbow Method
plt.plot(range(1, 11), Distortion)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()

# Choose the number of clusters based on the Elbow Method (e.g., 3 clusters)
num_clusters = 3

# Fit KMeans with the chosen number of clusters
kmeans = KMeans(n_clusters=num_clusters, init='k-means++', max_iter=300, n_init=10, random_state=0)
data['Cluster'] = kmeans.fit_predict(data[['CSUSHPISA', 'DATE_numeric']])

# Plot the scatterplot with clusters
sns.scatterplot(x='DATE_numeric', y='CSUSHPISA', data=data, hue='Cluster', palette='viridis')
plt.title('DATE_numeric vs CSUSHPISA')
plt.show()
```
